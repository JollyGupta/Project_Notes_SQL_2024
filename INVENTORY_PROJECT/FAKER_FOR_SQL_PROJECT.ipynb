```python
from faker import Faker
import numpy as np
import pandas as pd

# Initialize Faker
fake = Faker()

# Function to generate synthetic transaction data
def generate_synthetic_data(num_records):
    tranDate = []
    custName = []
    cardNum = []
    zipCode = []
    tranAmount = []

    # Generate synthetic data for each record
    for _ in range(num_records):
        tranDate.append(fake.date_time_between_dates('-1y', 'now'))  # Random transaction date within the last year
        custName.append(fake.name())  # Random customer name
        cardNum.append(fake.credit_card_number())  # Random credit card number
        zipCode.append(fake.zipcode())  # Random zip code
        tranAmount.append(np.random.randint(100, 5000))  # Random transaction amount between 100 and 5000

    # Create a DataFrame from the generated data
    data = {
        'Transaction Date': tranDate,
        'Customer Name': custName,
        'Card Number': cardNum,
        'Zip Code': zipCode,
        'Transaction Amount': tranAmount
    }

    # Convert to DataFrame
    df = pd.DataFrame(data)
    
    return df

# Generate synthetic data (you can change num_records to generate more records)
synthetic_data = generate_synthetic_data(100)  # 100 records for example

# Save data to CSV and Excel files
synthetic_data.to_csv('synthetic_data.csv', index=False)
synthetic_data.to_excel('synthetic_data.xlsx', index=False)

# Display the first few records of the synthetic data
synthetic_data.head(2)
```

```python
import pandas as pd
import random
from faker import Faker

# Initialize Faker
faker = Faker()

# Generate realistic data for each table
# Supplier Table
supplier_data = [
    {"SID": f"S{str(i).zfill(4)}",
     "SNAME": faker.company(), 
     "SADD": faker.address(), 
     "SCITY": faker.city(),
     "SPHONE": faker.phone_number(),
     "EMAIL": faker.email()}
    for i in range(1, 21)  # 20 suppliers
]

# Product Table
categories = ["IT", "HA", "HC"]
product_data = [
    {"PID": f"P{str(i).zfill(4)}",
     "PDESC": faker.bs(), 
     "PRICE": round(random.uniform(50, 500), 2),
     "CATEGORY": random.choice(categories),
     "SID": random.choice([s["SID"] for s in supplier_data])}
    for i in range(1, 51)  # 50 products
]

# Stock Table
stock_data = [
    {"PID": p["PID"],
     "SQTY": random.randint(5, 100),
     "ROL": random.randint(5, 15),
     "MOQ": random.randint(1, 10)}
    for p in product_data
]

# Customer Table
customer_data = [
    {"CID": f"C{str(i).zfill(4)}", 
     "CNAME": faker.name(),
     "ADDRESS": faker.address(), 
     "CITY": faker.city(),
     "PHONE": faker.phone_number(),
     "EMAIL": faker.email(), 
     "DOB": faker.date_of_birth(minimum_age=18, maximum_age=75)}
    for i in range(1, 31)  # 30 customers
]

# Orders Table
orders_data = [
    {"OID": f"O{str(i).zfill(4)}", 
     "ODATE": faker.date_this_year(), 
     "PID": random.choice([p["PID"] for p in product_data]), #A randomly selected Product ID from the product_data list.
     "CID": random.choice([c["CID"] for c in customer_data]), # A randomly selected Customer ID from the Customer_data list.
     "OQTY": random.randint(1, 10)}
    for i in range(1, 41)  # 40 orders
]

# Bill Table
#bill_data is created using a list comprehension. It takes data from the orders_data list and enriches it by pulling customer and product
#details from customer_data and product_data.Each dictionary in bill_data represents a bill for a specific order.

bill_data = [
    {
        "OID": o["OID"], 
        "ODATE": o["ODATE"], 
        "CNAME": next(c["CNAME"] for c in customer_data if c["CID"] == o["CID"]),#Matches the CID (Customer ID) in orders_data with the CID in customer_data.
        "ADDRESS": next(c["ADDRESS"] for c in customer_data if c["CID"] == o["CID"]),
        "PHONE": next(c["PHONE"] for c in customer_data if c["CID"] == o["CID"]), 
        "PDESC": next(p["PDESC"] for p in product_data if p["PID"] == o["PID"]),
        "PRICE": next(p["PRICE"] for p in product_data if p["PID"] == o["PID"]),
        "OQTY": o["OQTY"], "AMT": o["OQTY"] * next(p["PRICE"] for p in product_data if p["PID"] == o["PID"])
    }
    for o in orders_data # 40 order data 
]

# Purchase Table
purchase_data = [
    {"PID": p["PID"], 
     "SID": p["SID"],
     "PQTY": random.randint(10, 50),
     "DOP": faker.date_this_year()}
    for p in product_data[:30]  # Random subset of products
]

# Convert to DataFrames
supplier_df = pd.DataFrame(supplier_data)
product_df = pd.DataFrame(product_data)
stock_df = pd.DataFrame(stock_data)
customer_df = pd.DataFrame(customer_data)
orders_df = pd.DataFrame(orders_data)
bill_df = pd.DataFrame(bill_data)
purchase_df = pd.DataFrame(purchase_data)

# Save as CSV and Excel files
supplier_df.to_csv("/mnt/data/supplier.csv", index=False)
product_df.to_csv("/mnt/data/product.csv", index=False)
stock_df.to_csv("/mnt/data/stock.csv", index=False)
customer_df.to_csv("/mnt/data/customer.csv", index=False)
orders_df.to_csv("/mnt/data/orders.csv", index=False)
bill_df.to_csv("/mnt/data/bill.csv", index=False)
purchase_df.to_csv("/mnt/data/purchase.csv", index=False)

with pd.ExcelWriter("/mnt/data/tables.xlsx") as writer:
    supplier_df.to_excel(writer, sheet_name="Supplier", index=False)
    product_df.to_excel(writer, sheet_name="Product", index=False)
    stock_df.to_excel(writer, sheet_name="Stock", index=False)
    customer_df.to_excel(writer, sheet_name="Customer", index=False)
    orders_df.to_excel(writer, sheet_name="Orders", index=False)
    bill_df.to_excel(writer, sheet_name="Bill", index=False)
    purchase_df.to_excel(writer, sheet_name="Purchase", index=False)
```

```python
import pandas as pd
import random
from faker import Faker

# Initialize Faker
faker = Faker()

# Generate realistic data for each table
# Supplier Table
supplier_data = [
    {"SID": f"S{str(i).zfill(4)}",
     "SNAME": faker.company(), 
     "SADD": faker.address(), 
     "SCITY": faker.city(),
     "SPHONE": faker.phone_number(),
     "EMAIL": faker.email()}
    for i in range(1, 21)  # 20 suppliers
]

# Product Table
categories = ["IT", "HA", "HC"]
product_data = [
    {"PID": f"P{str(i).zfill(4)}",
     "PDESC": faker.bs(), 
     "PRICE": round(random.uniform(50, 500), 2),
     "CATEGORY": random.choice(categories),
     "SID": random.choice([s["SID"] for s in supplier_data])}
    for i in range(1, 51)  # 50 products
]

# Stock Table
stock_data = [
    {"PID": p["PID"],
     "SQTY": random.randint(5, 100),
     "ROL": random.randint(5, 15),
     "MOQ": random.randint(1, 10)}
    for p in product_data
]

# Customer Table
customer_data = [
    {"CID": f"C{str(i).zfill(4)}", 
     "CNAME": faker.name(),
     "ADDRESS": faker.address(), 
     "CITY": faker.city(),
     "PHONE": faker.phone_number(),
     "EMAIL": faker.email(), 
     "DOB": faker.date_of_birth(minimum_age=18, maximum_age=75)}
    for i in range(1, 31)  # 30 customers
]

# Orders Table
orders_data = [
    {"OID": f"O{str(i).zfill(4)}", 
     "ODATE": faker.date_this_year(), 
     "PID": random.choice([p["PID"] for p in product_data]),
     "CID": random.choice([c["CID"] for c in customer_data]),
     "OQTY": random.randint(1, 10)}
    for i in range(1, 41)  # 40 orders
]

# Convert to DataFrames
supplier_df = pd.DataFrame(supplier_data)
product_df = pd.DataFrame(product_data)
stock_df = pd.DataFrame(stock_data)
customer_df = pd.DataFrame(customer_data)
orders_df = pd.DataFrame(orders_data)

# Bill Table: Using Joins
# Step 1: Merge Orders with Customers
bill_df = orders_df.merge(customer_df, on="CID", how="left")

# Step 2: Merge the result with Products
bill_df = bill_df.merge(product_df, on="PID", how="left")

# Step 3: Calculate the Amount
bill_df["AMT"] = bill_df["OQTY"] * bill_df["PRICE"]

# Step 4: Keep relevant columns
bill_df = bill_df[["OID", "ODATE", "CNAME", "ADDRESS", "PHONE", "PDESC", "PRICE", "OQTY", "AMT"]]

# Purchase Table: Using Joins
# Step 1: Randomly select products for purchase
purchase_data = product_df.sample(n=30).merge(supplier_df, on="SID", how="left")
#how="left" ensures that this product will still appear in the result, 
#but the supplier columns (SNAME, SADD, etc.) will be filled with NaN for that row.

# Step 2: Add Purchase Quantity and Date
purchase_data["PQTY"] = [random.randint(10, 50) for _ in range(len(purchase_data))]
purchase_data["DOP"] = [faker.date_this_year() for _ in range(len(purchase_data))]

# Step 3: Keep relevant columns
#purchase_df = purchase_data[["PID", "SID", "PQTY", "DOP", "SNAME", "SADD", "SCITY", "SPHONE", "EMAIL"]]
purchase_df = purchase_df[["PID", "SID", "PQTY", "DOP"]]

# Save as CSV and Excel files
supplier_df.to_csv("supplier1.csv", index=False)
product_df.to_csv("product1.csv", index=False)
stock_df.to_csv("stock1.csv", index=False)
customer_df.to_csv("customer1.csv", index=False)
orders_df.to_csv("orders1.csv", index=False)
bill_df.to_csv("bill1.csv", index=False)
purchase_df.to_csv("purchase1.csv", index=False)

with pd.ExcelWriter("tables1.xlsx") as writer:
    supplier_df.to_excel(writer, sheet_name="Supplier1", index=False)
    product_df.to_excel(writer, sheet_name="Product1", index=False)
    stock_df.to_excel(writer, sheet_name="Stock1", index=False)
    customer_df.to_excel(writer, sheet_name="Customer1", index=False)
    orders_df.to_excel(writer, sheet_name="Orders1", index=False)
    bill_df.to_excel(writer, sheet_name="Bill1", index=False)
    purchase_df.to_excel(writer, sheet_name="Purchase1", index=False)
```
